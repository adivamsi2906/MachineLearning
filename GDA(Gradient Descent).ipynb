```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
```

```python
salarydata=pd.read_csv('C:/Users/adiva/Downloads/salary.csv')
```

```python
salarydata.head()
```

```python
x=np.array(salarydata['YearsExperience'])
y=np.array(salarydata['Salary'])
l=len(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=6)
lx=len(x_train)
```

```python
m=0.1
c=0.5
alpha=0.01
m_values=[]
c_values=[]
for i in range(4000):
    slope=0
    intercept=0
    for j in range(l):
        intercept=intercept+((m*x[j]+c)-y[j]) #updating gradient and intercepts until optimal
        slope=slope+((m*x[j]+c)-y[j])*x[j]
    c=c-((alpha*intercept)/l) #Updating weights and bias
    m=m-((alpha*slope)/l)
    c_values.append(c)
    m_values.append(m)
```

```python
print(f"slope is {m}")
print(f"intercept is {c}")
```

```python
plt.plot(m_values[:],c_values[:],marker='o',
         color='red',markerfacecolor='blue',
         markersize=11,linestyle='dashed')
plt.xlabel("slope")
plt.ylabel("intercept")
plt.title("convex function or bowl shape")
plt.show()
```

```python
y_pred=[]
for i in x_test:
    v=m*i+c #Y values for evey value of x in x_test
    y_pred.append(v)
print(y_pred)
```

```python
plt.plot(x_test,y_pred,marker='o',
        color='red',markerfacecolor='blue',
        markersize=11,linestyle='dashed')
plt.scatter(x,y,marker='o',color='blue')
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Linear Regression with one variable")
plt.show()
```
