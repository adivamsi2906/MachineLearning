```python
import numpy as np
x = np.random.random(15)
#creating random vector of size 15 with integers in range 1-20
x=np.random.randint(1,21,size=15)
print("Original array:")
print(x)
#reshaping the array to 3 by 5
x=x.reshape(3,5)
#printing array shape
print("The shape of the array is:")
print(x.shape)
print(x)
#replace max in each row by 0
x[np.arange(3),np.argmax(x,axis=1)]=0
print(x)
```

```python
#b) Eigenvalues and right eigenvectors
x1=np.array([[3,-2],[1,0]])
eigenval,eigenvector=np.linalg.eig(x1)
print("eigen values:",eigenval)
print("right eigen vectors:")
print(eigenvector)
```

```python
#c) sum of the diagonal element of a given array.
x2=np.array([[0,1,2],[3,4,5]])
diag_sum=np.trace(x2)
print("Diagnol sum:",diag_sum)
```

```python
#d)create a new shape to an array without changing its data.
x3=np.array([[1,2],[3,4],[5,6]])
reshape_x3=x3.reshape((2,3))
print("Reshaped 3x2 array")
print(reshape_x3)

x4=np.array([[1,2,3],[4,5,6]])
reshape_x4=x4.reshape((3,2))
print("Reshaped 2x3 array")
print(reshape_x4)
```
